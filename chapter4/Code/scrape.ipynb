# Now, do the loop across all links and store the results in a dataframe

Don't forget to store the data in a dataframe!

```python
all_date = pd.DataFrame()
missing = []

for j in tqdm(links):
    url = j
    
    try:
        data_1 = pd.read_html(url)[0]
        
        data_1 = data_1.drop([1, 3, 4], axis=1)
        data_1 = data_1.dropna()
        
        data_2 = pd.read_html(url)[2]
        data_2.columns = data_2.iloc[0]
        data_2 = data_2.drop(0)
        data_2 = data_2.drop('#', axis = 1)
        
        for i in range(len(data_1)):
            data_2[data_1.iloc[i][0]] = data_1.iloc[i][2]
            
        all_date = all_date.append(data_2, ignore_index = True)
        
    except:
        missing.append(j)
        next
        
    time.sleep(5)
    
    
    

   
        
```
